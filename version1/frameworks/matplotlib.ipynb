```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.array([1,2,5])
y = np.array([11,18,20])
```

```python
# Scatter Plot
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
# Plot
plt.plot(x,y,c="g")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Plot")
plt.show()
plt.savefig("Matplotlib_Plot.png")
```

```python
a = np.arange(0,5)
b = a * a
```

```python
# Subplot
plt.subplot(1,2,1)
plt.plot(a, b, "r--")

plt.subplot(1,2,2)
plt.plot(a,b,"g*--")
```

```python
# Bar Plot
sam_r15 = [10,20,30, 300]
sam_pulsar = [1,5,9]

plt.bar(sam_r15, sam_pulsar)
plt.xlabel("R15")
plt.ylabel("Pulsar")
plt.show()
```

```python
# Histogram plot

plt.hist(sam_pulsar)
```

```python
# Box plot
sam_r15 = [10,20,30, 60,70]
plt.boxplot(sam_r15,vert=True,patch_artist=True)
plt.show()
```

```python
# Pie Plot
sam_r15 = [10,20,30, 60]
labels = 'Python', 'C++', 'Ruby', 'Java'
explode = (0.4, 0.1, 0.1, 0.1)  # explode 1st slice
plt.pie(sam_r15, explode=explode, labels=labels)
```
